```python
import pandas as pd
import re
```

```python
file_path = r"C:\Users\romme\OneDrive\Desktop\pakwheels_cars.csv"
```

```python
data = pd.read_csv(file_path)
Dframe = pd.DataFrame(data)
```

```python
col = ["title", "price","city", "year", "mileage", "fuel_type", "engine_capacity", "transmission"]
```

```python
df2 = pd.DataFrame(data, columns=col)
df2.head()
```

```python
df2["year"] = df2["year"].astype('Int64')
```

```python
df2.head(1000)
```

```python
df2['title'] = (df2['title']
                .str.extract(r'^(.*?)(?=\s\d{4})', expand=False)
                .str.strip())

df2['price'] = (
             df2['price']
             .astype(str)
                .str.replace('PKR', '', regex=False)
                .str.replace(',', '', regex=False)
                .str.strip()
                )
df2['price'] = pd.to_numeric(df2['price'], errors='coerce')


df2['mileage'] = (
    df2['mileage']
        .astype(str)
        .str.replace('km', '', regex=False)
        .str.replace(',', '', regex=False)
        .str.strip()
)
df2['mileage'] = pd.to_numeric(df2['mileage'], errors='coerce')


df2['engine_capacity'] = (
    df2['engine_capacity']
        .astype(str)
        .str.replace('cc', '', regex=False)
        .str.replace('kWh', '', regex=False)
        .str.strip()
)
df2['engine_capacity'] = pd.to_numeric(df2['engine_capacity'], errors='coerce')

```

```python
df2 = df2[df2['fuel_type'] != 'Electric']
```

```python
df2.drop(df2[df2['fuel_type'] == 'Electric'].index, inplace=True)
```
